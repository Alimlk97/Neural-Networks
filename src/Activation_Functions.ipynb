```python
import math
import numpy as np
import matplotlib.pyplot as plt

# Define a range of values for x
x = np.linspace(-5, 5, 100)

# Sigmoid Activation Function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of Sigmoid Activation Function
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Hyperbolic Tangent (Tanh) Activation Function
def tanh(x):
    return np.tanh(x)

# Derivative of Tanh Activation Function
def tanh_derivative(x):
    return 1 - tanh(x) ** 2

# Rectified Linear Unit (ReLU) Activation Function
def relu(x):
    return np.maximum(0, x)

# Derivative of ReLU Activation Function
def relu_derivative(x):
    return np.where(x > 0, 1, 0)

# Create subplots for each activation function and its derivative
fig, axes = plt.subplots(2, 5, figsize=(15, 6))

# Sigmoid and its derivative
axes[0, 0].plot(x, sigmoid(x), label='Sigmoid')
axes[0, 0].set_title('Sigmoid')
axes[1, 0].plot(x, sigmoid_derivative(x), label='Sigmoid Derivative')
axes[1, 0].set_title('Sigmoid Derivative')

# Tanh and its derivative
axes[0, 1].plot(x, tanh(x), label='Tanh')
axes[0, 1].set_title('Tanh')
axes[1, 1].plot(x, tanh_derivative(x), label='Tanh Derivative')
axes[1, 1].set_title('Tanh Derivative')

# ReLU and its derivative
axes[0, 2].plot(x, relu(x), label='ReLU')
axes[0, 2].set_title('ReLU')
axes[1, 2].plot(x, relu_derivative(x), label='ReLU Derivative')
axes[1, 2].set_title('ReLU Derivative')

# Leaky ReLU and its derivative
alpha = 0.01
leaky_relu_func = lambda x: np.where(x > 0, x, alpha * x)
leaky_relu_derivative = lambda x: np.where(x > 0, 1, alpha)
axes[0, 3].plot(x, leaky_relu_func(x), label=f'Leaky ReLU (alpha={alpha})')
axes[0, 3].set_title(f'Leaky ReLU (alpha={alpha})')
axes[1, 3].plot(x, leaky_relu_derivative(x), label=f'Leaky ReLU Derivative (alpha={alpha})')
axes[1, 3].set_title(f'Leaky ReLU Derivative (alpha={alpha})')

# ELU and its derivative
alpha = 1.0
elu_func = lambda x: np.where(x > 0, x, alpha * (np.exp(x) - 1))
elu_derivative = lambda x: np.where(x > 0, 1, elu_func(x) + alpha)
axes[0, 4].plot(x, elu_func(x), label=f'ELU (alpha={alpha})')
axes[0, 4].set_title(f'ELU (alpha={alpha})')
axes[1, 4].plot(x, elu_derivative(x), label=f'ELU Derivative (alpha={alpha})')
axes[1, 4].set_title(f'ELU Derivative (alpha={alpha})')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

# Define a range of values for x
x = np.linspace(-5, 5, 100)

# Swish Activation Function
def swish(x):
    return x * sigmoid(x)

# Derivative of Swish Activation Function
def swish_derivative(x):
    return swish(x) + sigmoid(x) * (1 - swish(x))

# Scaled Exponential Linear Unit (SELU) Activation Function
def selu(x, alpha=1.67326, scale=1.0507):
    return scale * np.where(x > 0, x, alpha * (np.exp(x) - 1))

# Derivative of SELU Activation Function
def selu_derivative(x, alpha=1.67326, scale=1.0507):
    return scale * np.where(x > 0, 1, alpha * np.exp(x))

# Hard Swish Activation Function
def hard_swish(x):
    if isinstance(x, (int, float)):
        if x <= -3.0:
            return 0.0
        elif x >= 3.0:
            return x
        else:
            return x * (x + 3) / 6.0
    else:
        return np.array([0.0 if xi <= -3.0 else xi if xi >= 3.0 else xi * (xi + 3) / 6.0 for xi in x])

# Derivative of Hard Swish Activation Function
def hard_swish_derivative(x):
    if isinstance(x, (int, float)):
        if x <= -3.0 or x >= 3.0:
            return 0.0
        else:
            return (x + 3) / 6.0
    else:
        return np.array([0.0 if xi <= -3.0 or xi >= 3.0 else (xi + 3) / 6.0 for xi in x])

# Mish Activation Function
def mish(x):
    return x * np.tanh(np.log(1 + np.exp(x)))

# Derivative of Mish Activation Function
def mish_derivative(x):
    exp_x = np.exp(x)
    exp_2x = np.exp(2 * x)
    exp_x_2 = exp_x ** 2
    tanh_x = np.tanh(np.log(1 + exp_x))
    return exp_x * (4 * exp_2x + exp_x_2 + 4 * exp_x + exp_x_2 * tanh_x - 2 * exp_x * tanh_x) / ((exp_x_2 + 2 * exp_x * exp_2x + exp_x_2 * exp_2x) * (exp_x_2 + 2 * exp_x * exp_x_2 + exp_x_2 * exp_x_2))

# Create subplots for each activation function and its derivative
fig, axes = plt.subplots(2, 4, figsize=(15, 6))

# Swish and its derivative
axes[0, 0].plot(x, swish(x), label='Swish')
axes[0, 0].set_title('Swish')
axes[1, 0].plot(x, swish_derivative(x), label='Swish Derivative')
axes[1, 0].set_title('Swish Derivative')

# SELU and its derivative
alpha = 1.67326
scale = 1.0507
axes[0, 1].plot(x, selu(x, alpha, scale), label=f'SELU (alpha={alpha}, scale={scale})')
axes[0, 1].set_title(f'SELU (alpha={alpha}, scale={scale})')
axes[1, 1].plot(x, selu_derivative(x, alpha, scale), label=f'SELU Derivative (alpha={alpha}, scale={scale})')
axes[1, 1].set_title(f'SELU Derivative (alpha={alpha}, scale={scale})')

# Hard Swish and its derivative
axes[0, 2].plot(x, hard_swish(x), label='Hard Swish')
axes[0, 2].set_title('Hard Swish')
axes[1, 2].plot(x, hard_swish_derivative(x), label='Hard Swish Derivative')
axes[1, 2].set_title('Hard Swish Derivative')

# Mish and its derivative
axes[0, 3].plot(x, mish(x), label='Mish')
axes[0, 3].set_title('Mish')
axes[1, 3].plot(x, mish_derivative(x), label='Mish Derivative')
axes[1, 3].set_title('Mish Derivative')

# Adjust layout
plt.tight_layout()
plt.show()
```
